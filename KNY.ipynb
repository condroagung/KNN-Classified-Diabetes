```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

dataset = pd.read_csv('Diabetes.csv')
dataset
```

melakukan normalisasi data dari setiap field agar mendapatkan rentang nilai antara 0-1 kecuali field outcome karena field outcome merupakan klasifikasi kelas yang akan digunakan

```
dataset['Pregnancies'] = (dataset['Pregnancies']-dataset['Pregnancies'].min())/(dataset['Pregnancies'].max()-dataset['Pregnancies'].min())
dataset['Glucose'] = (dataset['Glucose']-dataset['Glucose'].min())/(dataset['Glucose'].max()-dataset['Glucose'].min())
dataset['BloodPressure'] = (dataset['BloodPressure']-dataset['BloodPressure'].min())/(dataset['BloodPressure'].max()-dataset['BloodPressure'].min())
dataset['SkinThickness'] = (dataset['SkinThickness']-dataset['SkinThickness'].min())/(dataset['SkinThickness'].max()-dataset['SkinThickness'].min())
dataset['Insulin'] = (dataset['Insulin']-dataset['Insulin'].min())/(dataset['Insulin'].max()-dataset['Insulin'].min())
dataset['BMI'] = (dataset['BMI']-dataset['BMI'].min())/(dataset['BMI'].max()-dataset['BMI'].min())
dataset['DiabetesPedigreeFunction'] = (dataset['DiabetesPedigreeFunction']-dataset['DiabetesPedigreeFunction'].min())/(dataset['DiabetesPedigreeFunction'].max()-dataset['DiabetesPedigreeFunction'].min())
dataset['Age'] = (dataset['Age']-dataset['Age'].min())/(dataset['Age'].max()-dataset['Age'].min())
dataset
```

fungsi melakukan pengoperasian KNN

```
def knn_operate(dataset_train,dataset_test,K,akurasi):
  for i in range(len(dataset_test)):

    dataset_train_copy = dataset_train.copy()
    dataset_train_copy

    dataset_train_copy['jarak'] = np.sqrt(pow(dataset_test.iloc[i,0] - dataset_train_copy['P'],2) + pow(dataset_test.iloc[i,1] - dataset_train_copy['G'],2) + pow(dataset_test.iloc[i,2] - dataset_train_copy['BP'],2)
    + pow(dataset_test.iloc[i,3] - dataset_train_copy['SK'],2) + pow(dataset_test.iloc[i,4] - dataset_train_copy['I'],2) + pow(dataset_test.iloc[i,5] - dataset_train_copy['BMI'],2)
    + pow(dataset_test.iloc[i,6] - dataset_train_copy['DP'],2) + pow(dataset_test.iloc[i,7] - dataset_train_copy['Age'],2))

    dataset_train_copy_sort = dataset_train_copy.sort_values(by ='jarak')

    count_1 = 0
    count_0 = 0
    count_k = 0

    for j in range(K):

      if (dataset_train_copy_sort.iloc[j,8] == 1):
        count_1 = count_1 + 1
      else:
        count_0 = count_0 + 1
  
    if (count_0 < count_1):
      count_k = 1
    else:
      count_k = 0
  
    if (count_k == dataset_test.iloc[i,8]):
      akurasi = akurasi + 1
    else:
      akurasi = akurasi + 0
  
  return akurasi,len(dataset_test)
```

```
K = 7
accurate = knn_operate(dataset1_train,dataset1_test,K,0)[0]
besar_data_set = knn_operate(dataset1_train,dataset1_test,K,0)[1]
accurate_1 = accurate / besar_data_set
print('Ketika K = ', K)
print('Dataset 1 benar :', accurate, 'dan salah : ', besar_data_set-accurate)
print(accurate_1*100,'%')
```
